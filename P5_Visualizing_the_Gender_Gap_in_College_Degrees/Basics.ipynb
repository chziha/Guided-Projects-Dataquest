```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt


women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(15, 20))


# Generate figures in the first column for STEM majors
# Subplots of the first column in 6*3, so we choose 1, 4, 7, 10, 13, and 16 for indexing

for sp in range(0, 18, 3):
    major_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[major_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[major_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
# Reduce the unnecessary y_ticks
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[major_index])
# Turn off all the x_axis labels to avoid overlap with titles
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
# Add a horizontal line acroos the figures to show degrees close to 50-50 gender breakdown 
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if major_index == 0:
        ax.text(2005, 10, "Men")
        ax.text(2003, 86, "Women")
    elif major_index == 5:
        ax.text(2005, 88, "Men")
        ax.text(2003, 7, "Women")
# Turn on the x_axis labels for the bottommost graph
        ax.tick_params(labelbottom='on')

# Similarly, generate figures in the second column for liberal arts majors

for sp in range(1, 16, 3):
    major_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[major_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[major_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[major_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if major_index == 0:
        ax.text(2005, 20, "Men")
        ax.text(2003, 78, "Women")
    elif major_index == 4:
        ax.tick_params(labelbottom='on')
    
# Generate figures in the third column for other majors

for sp in range(2, 20, 3):
    major_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[major_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[major_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[major_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if major_index == 0:
        ax.text(2005, 5, "Men")
        ax.text(2003, 92, "Women")
    elif major_index == 5:
        ax.text(2005, 65, "Men")
        ax.text(2003, 33, "Women")
        ax.tick_params(labelbottom='on')

import matplotlib
print(matplotlib.get_backend())

plt.savefig("gender_degrees.png")
plt.show()
```
