```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Check the basic information about the dataset
recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0], "\n")
print(recent_grads.head(), "\n")
print(recent_grads.tail(), "\n")
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads.dropna(axis = 0, inplace = True)
cleaned_data_count = recent_grads.shape[0]
print("before:", raw_data_count, " after:", cleaned_data_count)
```

```python
# Sample size and Median
ax1 = recent_grads.plot(x = "Sample_size", y = "Median", kind = "scatter", figsize = (5, 5))
ax1.set_title("Sample Size vs Median")
```

```python
# Sample size and umemployment rate
ax2 = recent_grads.plot(x = "Sample_size", y = "Unemployment_rate", kind = "scatter", figsize = (5, 5))
ax2.set_title("Sample size vs Unemployment rate")
```

```python
# Full time and median
ax3 = recent_grads.plot(x = "Full_time", y = "Median", kind = "scatter", figsize = (5, 5))
ax3.set_title("Full Time vs Median")
```

```python
# Share of women vs unemployment rate
ax4 = recent_grads.plot(x = "ShareWomen", y = "Unemployment_rate", kind = "scatter", figsize = (5, 5))
ax4.set_title("Share of Women vs Unemployment Rate")
```

```python
# Men vs median
ax5 = recent_grads.plot(x = "Men", y = "Median", kind = "scatter", figsize = (5, 5))
ax5.set_title("Men vs Median")
```

```python
# Women vs median
ax6 = recent_grads.plot(x = "Women", y = "Median", kind = "scatter", figsize = (5, 5))
ax6.set_title("Women vs Median")
```

```python
# Share of women vs median
ax7 = recent_grads.plot(x = "ShareWomen", y = "Median", kind = "scatter", figsize = (5, 5))
ax7.set_title("Share of Women vs Median")
```

From the plot of sample-size vs median, we can hardly see any correlation between these two parameters and the median doesn't become larger with a larger sample size, so students in more popular mahors do not make more money. From the plot of share of women vs median, we can see that with a higher of share of women in the subject, the median income seems to become lower, so students that majored in subjects that were majority female make less money. From the plot of full-time vs median, we can see that the correlation between these two parameters are very weak, we think there's no strong link between the number of full-time employees and median salary.
